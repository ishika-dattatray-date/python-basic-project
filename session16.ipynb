```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("insurance.csv")
```

```python
print(df.head(10))
```

```python
print("Shape:", df.shape)

print("\nInfo:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
numerical = df.select_dtypes(include=['int64','float64']).columns
categorical = df.select_dtypes(include=['object']).columns

print("Numerical Columns:")
print(numerical)

print("\nCategorical Columns:")
print(categorical)
```

```python
num_cols = ['age','bmi','children','expenses']

for col in num_cols:
    plt.figure(figsize=(5,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
cat_cols = ['sex','smoker','region']

for col in cat_cols:
    plt.figure(figsize=(5,4))
    sns.countplot(x=df[col])
    plt.title(f"Count Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Average Charges:", df['expenses'].mean())

print("Maximum Charges:", df['expenses'].max())

print("Minimum Charges:", df['expenses'].min())

print("\nAverage Charges by Smoker:")
print(df.groupby('smoker')['expenses'].mean())
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='smoker', y='expenses', data=df)
plt.title("expenses by Smoker")
plt.show()

plt.figure(figsize=(6,5))
sns.boxplot(x='sex', y='expenses', data=df)
plt.title("expenses by Sex")
plt.show()
```

```python
print("Average Age:", df['age'].mean())
print("Average BMI:", df['bmi'].mean())

print("\nRegion with Highest Customers:")
print(df['region'].value_counts())

print("\nSummary:")
print("1. Average age and BMI calculated.")
print("2. Smokers have much higher insurance charges than non-smokers.")
print("3. Region with maximum customers is shown above.")
print("4. Charges increase significantly for smokers.")
```
