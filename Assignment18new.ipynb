```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Step 2: Load dataset
df = pd.read_csv("ford_car_dataset.csv")

# Step 3: Handle missing values
print(df.isnull().sum())

# Step 4: Remove duplicate rows
df = df.drop_duplicates()

# Step 5: Histograms
df[['price', 'mileage', 'year', 'engineSize', 'mpg']].hist(figsize=(10,8))
plt.show()

# Step 6: Count Plots
sns.countplot(x='fuelType', data=df)
plt.show()

sns.countplot(x='transmission', data=df)
plt.show()

# Step 7: Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')
plt.show()

# Step 8: Input and Output Features
X = df.drop('price', axis=1)
y = df['price']

# Step 9: One-Hot Encoding
X = pd.get_dummies(X, drop_first=True)

# Step 10: Feature Scaling
scaler = StandardScaler()

numeric_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])

# Step 11: Display Final Dataset
X.head()
```
